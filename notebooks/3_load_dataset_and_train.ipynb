### Imports


```python
import numpy as np
import os
```

```python
def load_preprocessed_data(data_dir="./processed_data/crawled", combined=True):
    """Load preprocessed data from NPZ files.

    Args:
        data_dir: Directory containing preprocessed files
        combined: If True, load combined_dataset.npz; else load separate text/image files

    Returns:
        If combined: (text_features, image_features, labels)
        Else: (text_features, labels), (image_features, labels)
    """

    data_dir = os.path.expanduser(data_dir)

    if combined:
        # Load combined dataset
        combined_path = os.path.join(data_dir, "combined_dataset.npz")
        data = np.load(combined_path)
        return data["text_features"], data["image_features"], data["labels"]
    else:
        # Load separate files
        text_data = np.load(os.path.join(data_dir, "text_features.npz"))
        image_data = np.load(os.path.join(data_dir, "image_features.npz"))
        return (text_data["features"], text_data["labels"]), (
            image_data["features"],
            image_data["labels"],
        )
```

```python
loaded_data = load_preprocessed_data()
```
