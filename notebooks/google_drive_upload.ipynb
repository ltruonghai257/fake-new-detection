# Google Drive — Upload & Download

Standalone notebook for syncing large project files (JSON, images, trained models) to/from Google Drive.  
**Does not touch git** — use this alongside your normal `git push / git pull` workflow.

---

## Cross-machine workflow

```
Machine A (done working)          Machine B (starting fresh)
─────────────────────────────     ──────────────────────────────
1. git push                  →    1. git pull
2. Run "UPLOAD" cells below  →    2. Run "DOWNLOAD" cells below
```

**Auth options:**
- **Service account** (recommended) — create a key in Google Cloud, share the Drive folder with the service account email.
- **OAuth** (`USE_OAUTH = True`) — browser popup on first run; caches `token.pickle` next to the credentials file.

**One-time setup:**
1. `pip install -r requirements.txt` (from repo root)
2. Place your credentials JSON file somewhere safe (NOT committed to git — it's in `.gitignore` via `.env`)
3. Fill in the **Config** cell below

````python
import sys
import os

# Make src/ importable from this notebook
sys.path.insert(0, os.path.abspath("../src"))
````

## Config

Edit the values below — run this cell on **both** machines before any upload/download.

````python
# Path to your Google credentials JSON file.
# Can also be set via GOOGLE_APPLICATION_CREDENTIALS env var.
CREDENTIALS_PATH = "../service-account-key.json"  # <-- change this

# Set True if using an OAuth client JSON instead of a service account
USE_OAUTH = True

# Root folder on Google Drive — same name on both machines
DRIVE_ROOT_FOLDER = "fake-news-detection"
````

````python
from helpers.google_drive_uploader import GoogleDriveUploader

uploader = GoogleDriveUploader(
    credentials_path=CREDENTIALS_PATH,
    use_oauth=USE_OAUTH,
)
print("Authenticated successfully.")
````

---
# UPLOAD  (run on Machine A before switching machines)

````python
## U1. Upload JSON data
````

json_dir = "../data"  # <-- local directory containing JSON files

ids = uploader.upload_directory(
    local_dir=json_dir,
    folder_path=f"{DRIVE_ROOT_FOLDER}/data",
    extensions=[".json"],
    recursive=True,
    overwrite=True,
)
print(f"Uploaded {len(ids)} JSON file(s).")

````python
## U2. Upload images
````

image_dir = "../data/images"  # <-- local directory containing images

ids = uploader.upload_directory(
    local_dir=image_dir,
    folder_path=f"{DRIVE_ROOT_FOLDER}/images",
    extensions=[".jpg", ".jpeg", ".png", ".webp"],
    recursive=True,
    overwrite=True,
)
print(f"Uploaded {len(ids)} image(s).")

````python
## U3. Upload trained model files
````

model_dir = "../training/checkpoints"  # <-- directory with model checkpoints

ids = uploader.upload_directory(
    local_dir=model_dir,
    folder_path=f"{DRIVE_ROOT_FOLDER}/models",
    extensions=[".h5", ".hdf5", ".pkl", ".pt", ".pth", ".safetensors", ".npz", ".ckpt", ".bin"],
    recursive=True,
    overwrite=True,
)
print(f"Uploaded {len(ids)} model file(s).")

````python
---
# DOWNLOAD  (run on Machine B after `git pull`)

Run these cells to restore large files that are not tracked by git.
````

## D1. Download JSON data

````python
paths = uploader.download_folder(
    folder_path=f"{DRIVE_ROOT_FOLDER}/data",
    local_dir="../data",
    extensions=[".json"],
    recursive=True,
)
print(f"Downloaded {len(paths)} JSON file(s).")
````

## D2. Download images

````python
paths = uploader.download_folder(
    folder_path=f"{DRIVE_ROOT_FOLDER}/images",
    local_dir="../data/images",
    extensions=[".jpg", ".jpeg", ".png", ".webp"],
    recursive=True,
)
print(f"Downloaded {len(paths)} image(s).")
````

## D3. Download trained model files

````python
paths = uploader.download_folder(
    folder_path=f"{DRIVE_ROOT_FOLDER}/models",
    local_dir="../training/checkpoints",
    extensions=[".h5", ".hdf5", ".pkl", ".pt", ".pth", ".safetensors", ".npz", ".ckpt", ".bin"],
    recursive=True,
)
print(f"Downloaded {len(paths)} model file(s).")
````
